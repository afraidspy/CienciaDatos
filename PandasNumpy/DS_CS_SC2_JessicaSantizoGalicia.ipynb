c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base a como lo
reporta cada país en la columna vaccines)

```python
result= country_vaccinations.groupby('vaccines').count()
result
```

 d. Obtener la cantidad de vacunas aplicadas en todo el mundo. 

```python
total_vacunas = country_vaccinations['people_vaccinated'].sum()
total_vacunas
```

e. Calcular el promedio de vacunas aplicadas por país. 

```python
country_vaccinations[['iso_code','daily_vaccinations']].groupby('iso_code').mean()
```

```python
 promedio = country_vaccinations[['country','daily_vaccinations']].groupby('country').mean()
 promedio
```

f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo. 


```python
total_vacunas = country_vaccinations.loc[country_vaccinations['date'] == '2021-01-29']['total_vaccinations'].sum()
total_vacunas
```

g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  


```python
#country_vaccinations['diferencias'] = country_vaccinations['daily_vaccinations'] - country_vaccinations['daily_vaccinations_raw']
conDiferencias = country_vaccinations.assign(diferencias = country_vaccinations['daily_vaccinations'] - country_vaccinations['daily_vaccinations_raw'])
conDiferencias
```

h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua

```python
periodo = pd.to_datetime(country_vaccinations['date'].max()) - pd.to_datetime(country_vaccinations['date'].min())
periodo
```

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter /). 

```python
conCantidad  = country_vaccinations.assign(canVac = country_vaccinations['vaccines'].str.split('/ ').str.len())
conCantidad 
```

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020. 

```python

antes20 = country_vaccinations[pd.to_datetime(country_vaccinations['date']) < '2020-12-20']
antes20
```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer. 

```python
pfizer = country_vaccinations[country_vaccinations['vaccines'].str.contains('Pfizer',case=False)]
pfizer
```

l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter. 

```python
#Documentación: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.ExcelWriter.html
writer = pd.ExcelWriter('resultadosReto.xlsx')

conDiferencias.to_excel(writer, sheet_name='df_diferencias', index=False)
conCantidad.to_excel(writer, sheet_name='df_cantidad', index=False)
antes20.to_excel(writer, sheet_name='df_20-12-2020', index=False)
pfizer.to_excel(writer, sheet_name='df_Pfizer', index=False)
writer.save()
```
